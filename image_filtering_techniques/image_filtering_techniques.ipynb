```python
from google.colab.patches import cv2_imshow
import cv2
```

```python
# Read an image
img = cv2.imread('/content/MyDrive/MyDrive/cv.jpg')

# Display the original image
cv2_imshow(img)
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
cv2_imshow(gray)
```

```python
# Apply median blur with a kernel size of 3x3
median = cv2.medianBlur(gray, 3)

# Display the median-blurred image
cv2_imshow(median)
```

```python
# Apply a 3x3 mode filter to the grayscale image
mode = cv2.filter2D(gray, -1, 3)

# Display the mode-filtered image
cv2_imshow(mode)
```
