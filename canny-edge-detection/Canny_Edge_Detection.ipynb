```python
from google.colab.patches import cv2_imshow  # To display images in Colab
import cv2  # OpenCV library for image processing
```

```python
# Load an image from the specified path
img = cv2.imread('/content/MyDrive/MyDrive/cv.jpg')
cv2_imshow(img)  # Display the loaded image
```

```python
# Convert the loaded image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)  # Display the grayscale image
```

```python
# Load the same image as grayscale
segmented_img = cv2.imread("/content/MyDrive/MyDrive/cv.jpg", cv2.IMREAD_GRAYSCALE)
cv2_imshow(segmented_img)  # Display the segmented grayscale image
```

```python
# Apply Gaussian blur to the grayscale image
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Define lower and upper thresholds for the Canny edge detection
low_thres = 50
high_thres = 150

# Apply Canny edge detection to the blurred image
edges = cv2.Canny(blurred, low_thres, high_thres)
cv2_imshow(edges)  # Display the edges detected in the image
```
