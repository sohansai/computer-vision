```python
from google.colab.patches import cv2_imshow
import cv2
```

```python
## Read an image from the specified file path
img=cv2.imread('/content/MyDrive/MyDrive/one.jpg')
```

```python
# Display the loaded image
cv2_imshow(img)
```

```python
# Convert RGB to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
cv2_imshow(gray)
```

```python
# Read RGB values of a pixel
pixel_value = img[50, 100]

# Separate the pixel value into its B (Blue), G (Green), and R (Red) components
B, G, R = pixel_value

# Print the B, G, and R values of the pixel
print("Blue:", B)
print("Green:", G)
print("Red:", R)
```

```python
# Convert grayscale image to binary image
ret, thresh = cv2.threshold(gray, 70, 255, cv2.THRESH_BINARY)

# Display the binary image
cv2_imshow(thresh)
```

```python
# Get and print the shape of the image
print("Shape of the Image:", img.shape)  # Displays the dimensions (height, width, channels) of the image
```

```python
# Crop a region of interest (ROI) from the original image
img_crop = img[150:600, 200:900]

# Display the cropped ROI
cv2_imshow(img_crop)
```

```python
# Resize the image to a new width (200 pixels) and height (350 pixels) using linear interpolation.
img_resize = cv2.resize(img, (400, 350), interpolation=cv2.INTER_LINEAR)

# Display the resized image using cv2_imshow.
cv2_imshow(img_resize)
```

```python
# Rotate the image 180 degrees
img_rot = cv2.rotate(img, cv2.ROTATE_180)

# Display the rotated image
cv2_imshow(img_rot)
```

```python
import numpy as np

# Perform histogram equalization on the 'gray' image
hist = cv2.equalizeHist(gray)

# Horizontally stack the original 'gray' image and the equalized 'hist' image
res = np.hstack((gray, hist))

# Display the resulting image
cv2_imshow(res)
```
