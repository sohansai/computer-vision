```python
from google.colab.patches import cv2_imshow  # Import the OpenCV image display function
import cv2  # Import OpenCV library for image processing
```

```python
# Load the image from Google Drive
img = cv2.imread('/content/MyDrive/MyDrive/cv.jpg')
```

```python
# Display the original image
cv2_imshow(img)
```

```python
# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display the grayscale image
cv2_imshow(gray)
```

```python
import numpy as np

# Define the Laplacian high-pass filter kernel
laplacian = np.array([[0, 1, 0], [1, -4, 1], [0, 1, 0]])

# Apply the Laplacian filter to the grayscale image for sharpening
sharpened = cv2.filter2D(gray, -1, laplacian)

# Display the sharpened grayscale image
cv2_imshow(sharpened)
```

```python
# Convert the sharpened grayscale image back to color
sharpened_color = cv2.cvtColor(sharpened, cv2.COLOR_GRAY2BGR)

# Display the sharpened color image
cv2_imshow(sharpened_color)
```
